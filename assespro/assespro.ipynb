```python
numero_interaçoes = html_selector.xpath("//div[@class='et_pb_text_inner']").getall()
len(numero_interaçoes)
```

```python
nomes = []
certos = 0
errados_span = 0
errados_h5 = 0

for i in range(len(numero_interaçoes)):
    # Primeira tentativa com 'span/text()'
    try:
        nome_span = html_selector.xpath(f"//div[@class='et_pb_module et_pb_text et_pb_text_{i}  et_pb_text_align_left et_pb_bg_layout_light']/div[@class='et_pb_text_inner']/h5/span/text()").get()
        if nome_span is None:
            raise ValueError("Nome não encontrado no span.")
        else:
            certos += 1
            nomes.append(nome_span)
            print(nome_span)
            continue
    except Exception as e:
        errados_span += 1

    # Segunda tentativa com 'h5/text()'
    try:
        nome_h5 = html_selector.xpath(f"//div[@class='et_pb_module et_pb_text et_pb_text_{i}  et_pb_text_align_left et_pb_bg_layout_light']/div[@class='et_pb_text_inner']/h5/text()").get()
        if nome_h5 is None:
            print({i})
            raise ValueError("Nome não encontrado no h5.")
        else:
            certos += 1
            nomes.append(nome_h5)
            print(nome_h5)
            continue
    except Exception as e:
        errados_h5 += 1

    try:
        nome_clickabel = html_selector.xpath(f"//div[@class='et_pb_module et_pb_text et_pb_text_{i} et_clickable  et_pb_text_align_left et_pb_bg_layout_light']/div[@class='et_pb_text_inner']/h5/span/text()").get()
        if nome_clickabel is None:
            print({i})
            raise ValueError("Nome não encontrado no h5.")
        else:
            certos += 1
            nomes.append(nome_clickabel)
            print(nome_clickabel)
            continue
    except Exception as e:
        errados_h5 += 1

    try:
        nome_clickabel = html_selector.xpath(f"//div[@class='et_pb_module et_pb_text et_pb_text_{i} et_clickable  et_pb_text_align_left et_pb_bg_layout_light']/div[@class='et_pb_text_inner']/h5/span/text()").get()
        if nome_clickabel is None:
            print({i})
            raise ValueError("Nome não encontrado no h5.")
        else:
            certos += 1
            nomes.append(nome_clickabel)
            print(nome_clickabel)
            continue
    except Exception as e:
        errados_h5 += 1

print(f"{certos} certos")
print(f"{errados_span} span errados")
print(f"{errados_h5} span errados")
```
